## Tipos de datos.

```python
import numpy as np
```

Un array de enteros:

```python
arrInt = np.array([1,2,3,9])
arrInt.dtype, arrInt
```

Se puede definir el tipo de dato deseado para el array con el comando $dtype$:

```python
arrFloat = np.array([1,2,3,9], dtype='float64')
arrFloat.dtype, arrFloat
```

Se puede cambiar el tipo de dato del array:

```python
arr = np.array([0,1,2,3,4,5])
arr.dtype, arr
```

```python
arr = arr.astype(np.float64)
arr.dtype, arr
```

```python
arr = arr.astype(np.bool_)
arr.dtype, arr
```

```python
arr = arr.astype(np.str_)
arr.dtype, arr
```

El array solo puede tener un único tipo de dato.

## Dimensiones.

Un array de un solo valor tiene dimensión 0:

```python
scalar = np.array(2)
print(f"El array scalar = {scalar} tiene dimensión {scalar.ndim}.")
```

En numpy, un array que contiene una lista tiene dimensión 1:

```python
vector = np.array([1,2,3])
print(f"El vector = {vector} tiene dimensión {vector.ndim}.")
```

En numpy, un array que contiene una lista de listas tiene dimensión 2:

```python
matriz = np.array([[1,2,3],[4,5,6]])
print(f"La matriz = \n{matriz} \ntiene dimensión {matriz.ndim}.")
```

En numpy, los arrays son tensores de orden 0, 1, 2, 3, ... :

```python
tensor = np.array([[[1,0,0],[0,1,0],[0,0,1]],[[-1,0,0],[0,-1,0],[0,0,-1]]])
print(f"El tensor = \n{tensor}\ntiene orden {tensor.ndim}.")
```

Se puede cambiar la dimensión del array de numpy:

```python
vec = np.array([1,2,3],ndmin=4)
print(f"El vector vec = {vec} tiene dimensión {vec.ndim}.")
```

Se puede expandir la dimensión de un array en diferentes 'ejes':

```python
expand = np.expand_dims(np.array([1,2,3]),axis=0)
print(f"Diré entre comillas que el vector expand = \n{expand} \n'se convirtió en una matriz fila' y su dimensión en numpy es {expand.ndim}.")
expand = np.expand_dims(np.array([1,2,3]),axis=1)
print(f"De manera similar, el vector expand = \n{expand} \n'se convirtió en una matriz columna' y su dimensión en numpy es {expand.ndim}.")
```

Se puede también igualar la dimensión de un array con la de otro:

```python
matriz1 = np.array([[1,0],[0,1]])
vector1 = np.array([2,3])
print(f"Matriz1 = \n{matriz1}\nes una matriz y vector1 = {vector1} es un vector.")
vector1 = np.squeeze(matriz1)
print(f"Con la función squeeze se puede igualar ma dimensión del vector1 con la de la matriz1 obteniendo como resultado:\nvector1 = \n{vector1}")
```
