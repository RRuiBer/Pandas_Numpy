## Funciones Principales de Pandas

### Entorno de trabajo
Importamos pandas y realizamos una lectura de un archivo csv usando $pd.read\_ csv('/dirección/del/archivo.csv')$


```python
import pandas as pd
```

```python
data = pd.read_csv('data/bestsellers-with-categories.csv')
data.loc[:5]
```

### $.info()$
Muestra la información del contenido del data set.
$$data.info()$$

```python
data.info()
```

### $.describe()$
Muestra información estadística de las columnas numéricas.
$$data.describe()$$

```python
data.describe()
```

```python
data.shape[0]
```

### $.head()$
La función $head(n)$ muestra los primeros $n$ registros del data frame.
$$data.head(n),\hspace{0.4 cm}1\leq n \leq data.shape[0].$$
$data.shape[0]$ nos da el número de filas del data frame.

```python
data.head(5)
```

### $.tail()$
La función $tail(n)$ muestra los $n$ últimos registros del data frame.
$$data.tail(n),\hspace{0.4 cm}1\leq n \leq data.shape[0].$$
$data.shape[0]$ nos da el número de filas del data frame.

```python
data.tail(5)
```

### $.memory\_ usage()$
Muestra la memoria utilizada por el data frame.
$$data.memory\_ usaje(deep=True)$$

```python
data.memory_usage(deep=True)
```

### $.value\_ counts()$
Cuenta el número de coincidencias de una serie de datos en el data frame.
$$data['Columna'].value\_ counts()$$
En este ejemplo, de la $'Columna'$ especificada, muestra los datos contenidos y el número de veces que aparece cada dato.

```python
data['Author'].value_counts()
```

### Una opción para ingresar datos.

### $.drop\_ duplicates()$
Elimina datos duplicados.
$$data.drop\_ duplicates()$$

```python
data.drop_duplicates()
```

#### $keep=\ 'last'$
Dentro de la función $$data.drop\_ duplicate(keep=\ 'last')$$ $keep=\ 'last'$ elimina duplicados y mantiene el la ultima coincidencia.

```python
data.drop_duplicates(keep='last')
```

### $.sort\_ values()$
Con esta función podemos ordenar el contenido del data set por columnas:
$$data.sort\_ values('columna').$$
Dentro de esta función se puede indicar que el orden sea descendente usando la sentencia $ascending=False$.

```python
data.sort_values('Reviews', ascending=False)
```
