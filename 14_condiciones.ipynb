### Entorno de trabajo
Importamos pandas y realizamos una lectura de un archivo csv usando $pd.read\_ csv('/direcci√≥n/del/archivo.csv')$

```python
import pandas as pd
```

```python
data = pd.read_csv('data/bestsellers-with-categories.csv')
data.loc[:5]
```

### Explorar el data set
Aplicar condiciones al data frame genera datos booleanos. Podemos aplicar condiciones a las filas o a las columnas.

```python
data.loc[:10]['Year'] > 2012
```

Podemos reescribir la variable o crear una nueva en la que se apliquen las condiciones especificadas:

```python
data[data['Year'] > 2013]
```

Podemos filtrar datos de nuestro inter√©s ocupando las condiciones. Por ejemplo: libros de ficci√≥n publicados despu√©s del 2015.

```python
data[(data['Genre'] == 'Fiction') & (data['Year'] > 2015)]
```

En pandas, el s√≠mbolo ~ genera la negaci√≥n de la condici√≥n.

```python
data[(data['Genre'] == 'Fiction') & ~(data['Year'] > 2015)]
```
