## Leer archivos CSV y JSON con Pandas

```python
import pandas as pd
```

### CSV

Para leer archivos csv:
$$pd.read\_ csv('/dirección/del/archivo.csv')$$

```python
pd.read_csv('data/bestsellers-with-categories.csv')
```

Podemos agregar el separador como un parámetro extra de la función read_csv():
$$pd.read\_ csv('/dirección/del/archivo.csv',sep='|').$$
El separador por defecto es la coma, pero también puede ser el símbolo |.

En caso que el archivo no tenga especificado el header lo podemos asignar:
$$pd.read\_ csv('/dirección/del/archivo.csv',sep='|',header=i).\hspace{0.3 cm}\text{i es la fila que será el header}.$$

```python
data_books = pd.read_csv('/mnt/d/Users/Ricardo/Cursos/5_pandas_numpy/data/bestsellers-with-categories.csv')
```

Podemos explorar el archivo según sus filas y columnas:

```python
data_books.columns
```

También podemos editar el contenido por ejemplo, los nombres del header:
$$pd.read\_ csv('/dirección.csv',sep='|',header=i,\\ names=['Name', 'Author', 'User Rating', 'Reviews', 'Price', 'Year', 'Genre']).$$
El contenido de la lista name se puede editar.

## loc y iloc

Podemos explorar el documento por filas con la secuencia $[:]$.

```python
data_books[2:7]
```

También podemos explorar las columnas:

```python
data_books[['Author','Genre']]
```

### Loc

Permite explorar por filas y columnas.

```python
data_books.loc[0:4, ['Author','Genre']]
```

l
Loc permite operar con los datos de las casillas:

```python
data_books.loc[:, ['Price']]*0.5
```

También podemos filtrar los datos del archivo csv:

```python
data_books.loc[:,['Author']] == 'JJ Smith'
```

### Iloc
Permite explorar el archivo por filas y columnas por su índice mediante $:$.

```python
data_books.iloc[:,2:6]
```

```python
data_books.iloc[:2,2:]
```

Iloc también permite obtener valores del documento y operar con ellos:

```python
data_books.iloc[1,3]*2
```
