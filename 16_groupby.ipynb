## Groupby

### Entorno de trabajo
Importamos pandas y realizamos una lectura de un archivo csv usando $pd.read\_ csv('/dirección/del/archivo.csv')$

```python
import pandas as pd
```

```python
data = pd.read_csv('data/bestsellers-with-categories.csv')
data[:5]
```

### $.groupby()$


#### Ejemplo
Podemos contar los datos de una fila:
$$data.groupby('Columna').count()$$
Muestra el conteo de los datos contenidos en la fila indicada.

Agrupa los datos por autor y en cada columna muestra el número de coincidencias para el mismo autor (no muestra los datos referentes a la columna en sí).

```python
data.groupby('Author').count()
```

Podemos agrupar los datos conforme de diferentes formas:

```python
data.groupby(['Author','Year']).count()
```

```python
data.groupby('Author').min()
```

```python
data.groupby('Year').max()
```

```python
data.groupby('Year').min()
```

```python
data.groupby('Author').sum()
```

### $.agg()$
Método de agregación de datos.

```python
data.groupby('Author').agg(['min','max'])
```

```python
data.groupby('Author').agg({'Reviews':['min','max'],'User Rating':'sum'})
```
